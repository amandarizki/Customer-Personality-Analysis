## **Bagaimana struktur data dan pola distribusi dari variabel karakteristik pelanggan (usia, pendapatan, dan status keluarga)?**

Dalam analisis ini, langkah awal yang dilakukan adalah memahami karakteristik pelanggan melalui analisis struktur data dan pola distribusinya. Tiga variabel utama yang menjadi fokus dalam tahap ini adalah usia (Age; dihitung dari Year_Birth), tingkat pendidikan (Education), pendapatan (Income), dan status keluarga (Marital_Status, Kidhome, Teenhome).

Untuk memperoleh gambaran awal, dilakukan analisis deskriptif sederhana dengan menghitung nilai rata-rata, median, standar deviasi, serta ukuran bentuk distribusi seperti skewness dan kurtosis untuk masing-masing variabel.

### **Analisis Usia**

```python
df['Age'] = 2025 - df['Year_Birth']

# Central Tendency
print('Mean:',df['Age'].mean())
print('Median:',df['Age'].median())
print('Mode:',df['Age'].mode().values[0])

# Dispersion
print('Standard Deviation:',df['Age'].std())

# Skewness
print('Skewness:', df['Age'].skew())

# Kurtosis
print('Kurtosis:', df['Age'].kurtosis())
```

#### **Central Tendency**  
- **Rata-rata Usia**: **56,2 tahun**, menunjukkan mayoritas pelanggan berada di usia paruh baya.
- **Median**: **55 tahun**, yang berarti separuh pelanggan lebih muda dari 55 tahun dan separuh lainnya lebih tua.
- **Modus**: **49 tahun**, menunjukkan bahwa usia **49 tahun** adalah usia yang paling sering muncul di antara pelanggan.

#### **Penyebaran Data**  
- **Standard deviation** sebesar **11,98 tahun**, menunjukkan bahwa usia pelanggan tersebar cukup lebar di sekitar rata-rata, dengan sebagian besar pelanggan berada dalam rentang usia yang relatif seimbang.

#### **Skewness**  
- **Skewness**: **0,35**, yang menunjukkan distribusi usia sedikit condong ke kanan (positif), artinya ada lebih banyak pelanggan yang lebih muda dibandingkan dengan yang lebih tua, meskipun perbedaan ini tidak terlalu signifikan.

#### **Kurtosis**  
- **Kurtosis**: **0,72**, menunjukkan distribusi usia relatif datar dan tidak menunjukkan adanya outlier ekstrem yang signifikan.

---

#### **Insight**  
Mayoritas pelanggan berada di usia **paruh baya**, dengan konsentrasi usia sekitar **49-56 tahun**. Dengan distribusi yang cukup merata, strategi pemasaran dapat difokuskan pada segmen usia ini.

Berdasarkan output di atas, terlihat bahwa pada kolom Marital_Status, terdapat banyak kategori. Pada analisis ini, kita hanya memerlukan perbandingan untuk pelanggan yang sudah menikah dan yang belum/tidak menikah. Oleh karena itu, kita akan melakukan pengelompokan menjadi 2 kategori sebelum analisis. Dimana, untuk pengelompokannya:
1. Married: Married, Together
2. Not Married: Single, Divorced, Widow, Alone, Absurd, YOLO

```python
df['MaritalGroup'] = df['Marital_Status'].apply(lambda x: 'Married' if x in ['Married', 'Together'] else 'Not Married')
df['MaritalGroup'].value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='MaritalGroup', y='MntWines', data=df, estimator=np.mean, ci=None)
plt.title('Rata-rata Pengeluaran Wine antara Menikah dan Tidak Menikah')
plt.xlabel('Status Pernikahan')
plt.ylabel('Rata-rata Pengeluaran Wine')
plt.show()
```

Perbandingan Pengeluaran Wine antara Menikah dan Tidak Menikah:

Orang yang tidak menikah (Not Married) memiliki sedikit lebih tinggi rata-rata pengeluaran untuk wine dibandingkan dengan yang menikah (Married).

Perbedaannya tidak terlalu signifikan secara visual, namun tetap menunjukkan bahwa individu yang tidak menikah cenderung mengalokasikan lebih banyak pengeluaran untuk wine.

---

**Kesimpulan**

Tidak tampak perbedaan mencolok secara visual antara pelanggan menikah dan tidak menikah dalam hal pengeluaran wine. Akan tetapi, untuk mengetahui apakah perbedaannya signifikan secara statistik, kita perlu melanjutkan ke Two Sample T-Test.

```python
married_wine = df[df['MaritalGroup'] == 'Married']['MntWines']
not_married_wine = df[df['MaritalGroup'] == 'Not Married']['MntWines']

t_stat, p_value = ttest_ind(married_wine, not_married_wine, equal_var=False)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretasi otomatis
if p_value < 0.05:
    print("Terdapat perbedaan signifikan rata-rata pengeluaran wine antara pelanggan yang menikah dan belum menikah.")
else:
    print("Tidak terdapat perbedaan signifikan rata-rata pengeluaran wine antara pelanggan yang menikah dan belum menikah.")
```

**Hipotesis**

**Hipotesis nol (H₀):** Tidak ada perbedaan signifikan rata-rata pengeluaran wine antara pelanggan yang menikah dan belum menikah. Artinya, pengeluaran wine antara pelanggan yang menikah dan belum menikah memiliki rata-rata yang sama.

**Hipotesis alternatif (H₁):** Ada perbedaan signifikan rata-rata pengeluaran wine antara pelanggan yang menikah dan belum menikah. Artinya,  pengeluaran wine antara pelanggan yang menikah dan belum menikah memiliki rata-rata yang berbeda.


---


**Taraf Signifikansi**

Alfa = 5%


---


**Statistik Uji**

Berdasarkan hasil uji Independent t-test, didapatkan:

**T-statistic: -0.28**

**P-value: 0.7758**


---


**Keputusan**

Dengan nilai p-value > 0.05, maka gagal menolak hipotesis nol (H₀). Atau dengan kata lain hipotesis nol (H₀) diterima.

---

**Kesimpulan**

Artinya, tidak ada perbedaan signifikan rata-rata pengeluaran wine antara pelanggan yang menikah dan belum menikah.
